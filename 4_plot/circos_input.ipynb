```python
import os
import pandas as pd 
import numpy as np 
```

```python
PATH='C:/Users/houy2/Documents/Yuan/projects/AD/AD_genetics/AD_MR'
input_DT="DT"
input_raw="DT/raw"
input_circos='DT/circos'
input_kar='DT/circos/karyotype'
output_chr='DT/circos/chr'
output_chr_pos="DT/circos/chr_pos"
output_circos="DT/circos/mr_circos_v9"
datetime="Dec18"
os.chdir(PATH)
```

```python
df = pd.read_csv("./%s/brain_MR_alltargets_#1176_%s.tsv" % (input_DT,datetime),sep="\t")
sub_q=df[df["sig#_FDR"]>0].copy()
sub_q["FDR_sig"]='yes'
sub_q["type"]='FDR'
sub_q=sub_q[['symbol','type','FDR_sig']]
print("FDR #%s" % sub_q.shape[0])
sub_p=df[df["sig#"]>0].copy()
sub_p["pvalue_sig"]='yes'
sub_p["type"]='pval'
sub_p=sub_p[['symbol','type','pvalue_sig']]
print("P #%s" % sub_p.shape[0])
sub=pd.concat([sub_q,sub_p])
sub=sub.drop_duplicates(["symbol"])
sub_n=df[['symbol','alzMR score']]
sub=pd.merge(sub,sub_n,how="outer",on="symbol")
print(sub.shape)
```

```python
#Please load a label file including gene symbol generate in alzMR score.
df_label = pd.read_csv("./%s/MR_highlight#95_%s.tsv" % (input_DT,datetime),sep="\t")
df_label=df_label[["symbol","Race_group","highlight"]].copy()
df_label["label"]='yes'
```

```python
df_merge=pd.merge(df_label,sub,how="outer",on="symbol")
###add label after file generate
df_merge.to_csv("./%s/MR_allTargets_#%s_label_%s.tsv" % (input_circos,sub.shape[0],datetime), sep='\t',index=False)
```

```python
def unique(list1): 
    x = np.array(list1) 
    return np.unique(x)
```

```python
df = df_merge.copy()
print(df.shape)
```

```python
df_gene = pd.read_csv("./%s/gencode_v39lift37_protein_coding_gene.tsv" % input_kar,sep="\t")
print(df_gene.shape)
df_gene.head()
```

```python
df_kary = pd.read_csv("./%s/karyotype.human.hg19_mr.tsv" % input_kar,sep="\t")
print(df_kary.shape)
df_kary
```

```python
df_gene = pd.merge(df,df_gene,how = "inner",on =["symbol"])
df_gene=df_gene.sort_values(by='start', ascending=True)
print(df_gene.shape)
df_gene = pd.merge(df_kary,df_gene,how = "inner",on =["chr"])
df_gene=df_gene[["tag","icon","id","start","end","color","symbol","label","FDR_sig","pvalue_sig"]]
print(df_gene.shape)
df_gene.head()
```

```python
df_gene.to_csv("./%s/MR_allTargets_#%s_pos.tsv" % (input_circos,df_gene.shape[0]), sep='\t',index=False)
```

```python
chr_list=unique(df_gene['id'].tolist())
for c in chr_list:
   sub=df_gene[df_gene["id"]==c]
   sub.to_csv("./%s/MR_allTargets_%s.tsv" % (output_chr,c), sep='\t',index=False)
```

```python
listfile_chr = os.listdir("./%s" % output_chr)
for file in listfile_chr:
   chr_file=open("./%s/%s" % (output_chr,file),"r")
   line1=chr_file.readline()
   L1=line1.strip("\n").split("\t")
   df_pos=open("./%s/%s" % (output_chr_pos,file),"w")
   df_pos.write("tag\ticon\tid\tstart\tend\tcolor\tsymbol\n")
   S=0
   for line in chr_file.readlines():
      L=line.strip('\n').split('\t')
      fdr=L[L1.index("FDR_sig")]
      pval=L[L1.index("pvalue_sig")]
      color=L[L1.index("color")]
      gene=L[L1.index("symbol")]
      if fdr =="yes":
         start=S
         end=S+3
         S+=3
      elif pval=="yes":
         start=S
         end=S+2
         S+=2
      else:
         start=S
         end=S+1
         S+=1
      df_pos.write("%s\t%s\t%s\t%s\t%s\t%s\t%s\n" % (L[0],L[1],L[2],start,end,color,gene))
   chr_file.close()
   df_pos.close()
```

```python
list_hs = ["hs1","hs2","hs3","hs4","hs5","hs6","hs7","hs8","hs9","hs10","hs11","hs12","hs13","hs14","hs15","hs16","hs17","hs18","hs19","hs20","hs21","hs22"]
df_chr = pd.DataFrame()
id=[]
start=[]
end=[]
for f in list_hs:
   df_hs=pd.read_csv("./%s/MR_allTargets_%s.tsv" % (output_chr_pos,f), sep="\t")
   df_chr = pd.concat([df_hs, df_chr])
   id.append(f)
   start.append(df_hs["start"].tolist()[0])
   end.append(df_hs["end"].tolist()[-1])
print(df_chr.shape)
df_chr.head()
```

```python
df_kary = pd.read_csv("./%s/karyotype.human.hg19_mr.tsv" % input_kar,sep="\t")
df_kary_sub=df_kary[["id","order"]]
df_chr = pd.merge(df_kary_sub,df_chr,how = "inner",on =["id"])
df_chr=df_chr[["id","start","end","symbol"]]
df_chr.to_csv("./%s/karyotype.human.hg19_%s.tsv" % (input_kar,df_chr.shape[0]),index=False,sep='\t')
df_chr.head()
```

```python
Karyotype = pd.DataFrame.from_dict({'id':id,'start':start,'end':end})
Karyotype = pd.merge(df_kary,Karyotype,how = "inner",on =["id"])
Karyotype=Karyotype[["tag","icon","id","order","start","end","color"]]
Karyotype.to_csv("./%s/karyotype.human.hg19_%s_%s.txt" % (input_kar,df_chr.shape[0],datetime),index=False,sep=' ',header=False)
Karyotype
```

```python
W=0.5
df_CHR=open("./%s/karyotype.human.hg19_%s.tsv" % (input_kar,df_chr.shape[0]),"r")
df_CHR.readline()
df_win=open("./%s/karyotype.human.hg19_%s_0.5.tsv" % (input_kar,df_chr.shape[0]),"w")
df_win.write("id\tstart\tend\tsymbol\n")
for line in df_CHR.readlines():
    L=line.strip('\n').split('\t')
    middle_value=float(L[1])+float((float(L[2])-float(L[1]))/2)
    if middle_value-W <= 0:
        df_win.write("%s\t0\t%d\t%s\n" % (L[0], float(middle_value+W),L[3]))
    else:
        df_win.write("%s\t%d\t%d\t%s\n" % (L[0], float(middle_value-W), float(middle_value+W),L[3]))
df_CHR.close()
df_win.close()
```

```python
df_DT=pd.read_csv("./%s/DT_IVW_500kb_Sep13_#29027.tsv" % input_raw, sep="\t")
print(df_DT.shape)
GWAS_files=unique(df_DT["Outcome"].tolist())
for AD in GWAS_files:
    sub_AD=df_DT[df_DT["Outcome"]==AD].copy().reindex()
    for i in sub_AD.index:
        if sub_AD.loc[i,"FDR"] < 0.05:
            sub_AD.loc[i,"sig"]="id=FDR"
        elif sub_AD.loc[i,"Pvalue"] < 0.05:
            sub_AD.loc[i,"sig"]="id=pval"
        else:
            sub_AD.loc[i,"sig"]="id=non_sig"

    for i in sub_AD.index:
        if sub_AD.loc[i,"xQTL_type"] =='Cortex-ROSMAP.All_pQTL':
            sub_AD.loc[i,"pos"]=12
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-ROSMAP.Ctl_pQTL':
            sub_AD.loc[i,"pos"]=11
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-Banner_pQTL':
            sub_AD.loc[i,"pos"]=10
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-ROSMAP_eQTL':
            sub_AD.loc[i,"pos"]=9
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-Mayo_eQTL':
            sub_AD.loc[i,"pos"]=8
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-Metabrain_eQTL':
            sub_AD.loc[i,"pos"]=7
        elif sub_AD.loc[i,"xQTL_type"] =='Cortex-Meta_eQTL':
            sub_AD.loc[i,"pos"]=6
        elif sub_AD.loc[i,"xQTL_type"] =='Cerebellum-Mayo_eQTL':
            sub_AD.loc[i,"pos"]=5
        elif sub_AD.loc[i,"xQTL_type"] =='Cerebellum-Metabrain_eQTL':
            sub_AD.loc[i,"pos"]=4
        elif sub_AD.loc[i,"xQTL_type"] =='BasalGanglia-Metabrain_eQTL':
            sub_AD.loc[i,"pos"]=3
        elif sub_AD.loc[i,"xQTL_type"] =='Hippocampus-Metabrain_eQTL':
            sub_AD.loc[i,"pos"]=2
        elif sub_AD.loc[i,"xQTL_type"] =='SpinalCord-Metabrain_eQTL':
            sub_AD.loc[i,"pos"]=1
 
    sub_AD_rename=sub_AD.rename(columns={"Exposure":"symbol"})
    sub_AD_chr=pd.merge(df_chr,sub_AD_rename, how="inner",on="symbol")
    sub_AD_chr=sub_AD_chr[["id","start","end","pos","sig"]]
    print("%s #%s" % (AD,sub_AD_chr.shape[0]))
    sub_AD_chr.to_csv("./%s/%s_all.txt" % (output_circos,AD), sep=" ",index=False,header=False)
    print("==========")
```
